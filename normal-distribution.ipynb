# Distribución normal

Es una distribución de probabilidad de Variable Continua. La gráfica de su función de densidad tiene forma de campana y es simétrica respecto a $\mu$

Esta distribución permite modelar muchos fenómenos naturales. Algunos de ellos son:
- Caracteres de individuos (estatura, IQ)
- Caracteres fisiológicos de los efectos de un fármaco
- Caracteres sociologicos como el consumo de un producto
- Nivel de ruido en telecomunicaciones
- Errores de medición (gracias a la función error de gauss)

También aparece en áreas propoias de la estadística:

- La distribución muestral de las medias muestrales es aproximadamente normal aunque la distribución poblacional no lo sea.

El último si se lee rápido es un gran WTF pero lo que quiere decir es que si tomamos muchas muestras aleatorias de una población más grande, la media de esas muestras tienden a distribuirse de forma normal.

### Propiedades

- Simétrica respecto a $\mu$
- La moda y la mediana son iguales a $\mu$
- en el intervalo [$\mu - \sigma;\mu+\sigma$] está comprendida el 68% de la distribución


Algo muy interesante que toda persona que trabaja con datos debe manejar es el teorema del límite central

### Teorema del Límite Central

La suma de un gran número de variables aleatorias se distribuye aproximadamente como una normal.
La importancia de este teorema es que la función de distribución de la normal puede usarse para aproximar otras funciones de distribucion



- El teorema del límite central garantiza una distribución aproximadamente normal cuando n es suficientemente grande.
- Establece que es suficiente que las variables que se suman sean independientes, idénticamente distribuidas, con valor esperado y varianza finitas.
